```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from descartes import PolygonPatch
from shapely.geometry import Point, Polygon, MultiPolygon
```

```python
w, h = 2, 2
west, east, south, north = 0, 100, 100, 0
grid = gpd.GeoDataFrame()
for row in range(h, 100+h, h):
    for col in range(w, 100+w, w):
        poly = Polygon(((col-w, row), (col, row), (col, row-h), 
                       (col-w, row-h),(col-w, row)))
        temp = gpd.GeoDataFrame({'geometry': [poly]})
        grid = grid.append(temp)
grid.reset_index(drop=True, inplace=True)
grid.head()
```

```python
x = np.random.uniform(1, 100, 10000)
y = np.random.uniform(1, 100, 10000)

points = gpd.GeoDataFrame(data={'x':x, 'y':y})
points['geometry'] = points.apply(lambda row: Point((row['x'], row['y'])), axis=1)
rtree_index = points.sindex
points.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6.5))

for i, g in grid['geometry'].iteritems():
    patch = PolygonPatch(g, fc='#dcdcdc', ec='r', zorder=2)
    patch2 = PolygonPatch(g, fc='#dcdcdc', ec='r', zorder=2)
    ax[0].add_patch(patch)
    ax[1].add_patch(patch2)

ax[0].set_xlim(west, east)
ax[0].set_ylim(south, north)
ax[1].set_xlim(west, east)
ax[1].set_ylim(south, north)

ax[1].scatter(points.iloc[:, 0].values, points.iloc[:, 1].values, 
              zorder=3, marker='.', color='b')
```

```python
within = pd.DataFrame()
poly = grid['geometry'].iloc[18]

possible_matches_index = list(rtree_index.intersection(poly.bounds))
possible_matches = points.iloc[possible_matches_index]
precise_matches = possible_matches[possible_matches.intersects(poly)]
within = within.append(precise_matches)

within = within.drop_duplicates(subset=['x', 'y'])
outside = points[~points.isin(within)]


fig, ax = plt.subplots(1, 2, figsize=(15, 6.5))
for i, g in grid['geometry'].iteritems():
    patch = PolygonPatch(g, fc='#dcdcdc', ec='r', zorder=2)
    patch2 = PolygonPatch(g, fc='#dcdcdc', ec='r', zorder=2)
    ax[0].add_patch(patch)
    ax[1].add_patch(patch2)

ax[0].scatter(points.iloc[:, 0].values, points.iloc[:, 1].values, 
              zorder=3, marker='.', color='b')

ax[1].scatter(x=outside['x'], y=outside['y'], 
           c='#cdcdcd', linewidth=0, zorder=3, alpha=.5, marker='.')
ax[1].scatter(x=within['x'], y=within['y'], 
           c='r', linewidth=0, zorder=3, marker='.')

ax[0].set_xlim(west, east)
ax[0].set_ylim(south, north)
ax[1].set_xlim(west, east)
ax[1].set_ylim(south, north)
```
